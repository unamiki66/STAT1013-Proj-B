```python
import numpy as np

t_value = (120.37 - 121) / (23.27 / np.sqrt(250))
print('test statistic is: %.3f' %t_value)
```

```python
from scipy.stats import t

T = t(df=250)
## alternative is two-sided
## p = 2P(T > |t|) = 2*(1 - F(|t|))
p = 2*(1.0 - T.cdf(abs(t_value)))

print('p-value is %f' %p)
```
